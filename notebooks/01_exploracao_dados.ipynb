```python
# =====================================
# 01 - Exploração Inicial dos Dados
# =====================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', None)

# Carregamento dos dados
# O erro FileNotFoundError indica que o arquivo não foi encontrado no caminho especificado.
# Por favor, verifique se o arquivo 'amazon_delivery_dados_alunos.csv' está no local correto
# (ex: na pasta 'data/raw' dentro do seu ambiente Colab, ou diretamente no diretório raiz).
# Ajustando o caminho para um formato comum no Colab. Se o arquivo estiver em outro local,
# você precisará ajustar o caminho novamente.
df = pd.read_csv('amazon_delivery_dados_alunos.csv') # Alterado para carregar da raiz

# Visualização inicial
df.head()
```

```python

# Estrutura do dataset
df.info()
```

```python

# Dimensão do dataset
print(f'Linhas: {df.shape[0]}')
print(f'Colunas: {df.shape[1]}')
```

```python

# Valores ausentes
df.isna().sum()
```

```python

# Conversão de tipos numéricos
cols_numericas = [
    'Agent_Rating',
    'Store_Latitude', 'Store_Longitude',
    'Drop_Latitude', 'Drop_Longitude'
]

for col in cols_numericas:
    df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python

# Conversão de datas
df['Order_Date'] = pd.to_datetime(df['Order_Date'])
```

```python

# Estatísticas descritivas
df[['Agent_Age', 'Agent_Rating', 'Delivery_Time']].describe()
```
